# Getting Rid of Unnecessary Columns Before EDA and Training

```python
import pandas as pd
import os

output_dir = '../../Data/Warehouse/MergedData'
output_file = output_dir + '/merged_data_clean.csv'

# Load in our data
filepath = output_file
data = pd.read_csv(filepath)

data.head(-5)
```

```python
cols_to_keep = ['POSITION', 'PLAYER', 'TEAM', 'GP', 'HANDED', 'AAV', 'SEASON', 'SALARY CAP PERCENTAGE', 'SALARY CAP']

# List all columns that do not end with '/GP' or '/60'
all_columns_to_drop = [col for col in data.columns if not col.endswith('/GP') and not col.endswith('/60') and not col.endswith('%') and col not in cols_to_keep]

data = data.drop(columns=all_columns_to_drop)
print(data.shape, data.columns.to_list())
```

```python
# Extract all columns that end in '/60' or '/GP' or '%'
per_60_columns = data.filter(like='/60')
per_GP_columns = data.filter(like='/GP')
percentage_columns = data.filter(like='%')

# Extract all columns that are not per 60 or per GP
non_per_columns = data.drop(per_60_columns.columns, axis=1).drop(per_GP_columns.columns, axis=1).drop(percentage_columns.columns, axis=1)

final_data = pd.concat([non_per_columns, per_GP_columns, per_60_columns, percentage_columns], axis=1)
final_data.columns.tolist()
```

```python
custom_chosen_features = ['TEAM',
 'PLAYER',
 'POSITION',
 'GP',
 'HANDED',
 'AAV',
 'SEASON',
 'SALARY CAP',
 'SALARY CAP PERCENTAGE',
 'GOALS/GP',
 'TOTAL ASSISTS/GP',
 'FIRST ASSISTS/GP',
 'SECOND ASSISTS/GP',
 'TOTAL POINTS/GP',
 'SHOTS/GP',
 'IXG/GP',
 'ICF/GP',
 'IFF/GP',
 'ISCF/GP',
 'IHDCF/GP',
 'RUSH ATTEMPTS/GP',
 'REBOUNDS CREATED/GP',
 'PENALTIES DRAWN/GP',
 'GIVEAWAYS/GP',
 'TAKEAWAYS/GP',
 'HITS/GP',
 'HITS TAKEN/GP',
 'SHOTS BLOCKED/GP',
 'FACEOFFS WON/GP',
 'FACEOFFS LOST/GP',
 'CF/GP',
 'CA/GP',
 'FF/GP',
 'FA/GP',
 'SF/GP',
 'SA/GP',
 'GF/GP',
 'GA/GP',
 'XGF/GP',
 'XGA/GP',
 'SCF/GP',
 'SCA/GP',
 'HDCF/GP',
 'HDCA/GP',
 'HDGF/GP',
 'HDGA/GP',
 'MDCF/GP',
 'MDCA/GP',
 'MDGF/GP',
 'MDGA/GP',
 'LDCF/GP',
 'LDCA/GP',
 'LDGF/GP',
 'LDGA/GP']

final_data = final_data[custom_chosen_features]
```

# Saving Final Changes

```python
# Save the merged data to a csv file
output_dir = '../../Data/Warehouse/MergedData'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

output_file = output_dir + '/ML_data.csv'
final_data.to_csv(output_file, index=False)
```
